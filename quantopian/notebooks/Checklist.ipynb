```python
## import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from quantopian.pipeline import Pipeline, CustomFactor
from quantopian.pipeline.data.builtin import USEquityPricing
from quantopian.pipeline.filters import QTradableStocksUS, StaticAssets
from quantopian.pipeline.classifiers.fundamentals import Sector
from quantopian.pipeline.data import morningstar, Fundamentals as F
from quantopian.research import run_pipeline
from quantopian.pipeline.factors import SimpleMovingAverage

      
BILLION = 1000000000


def make_pipeline():
    mask = StaticAssets(symbols([
        #'M',
        #'BSET',
       # 'SNAP',
        #'AXP',
        #'JPM',
        #'SQ',
        'V'
    ]))
    
    revenue_growth = F.revenue_growth.latest
    earnings_growth = F.net_income_growth.latest
    bps_growth = F.equity_per_share_growth.latest
    dps_growth = F.dps_growth.latest
    
    net_margin = F.net_margin.latest
    
    roe = F.roe.latest
    roic = F.roic.latest
    cash = F.cash_and_cash_equivalents.latest
    op_income = F.operating_income.latest

    debt_equity = F.total_debt_equity_ratio.latest
    current_ratio = F.current_ratio.latest
    
    current_assets = F.current_assets.latest
    total_debt = F.total_debt.latest
    
    total_liabilities = F.total_liabilities_net_minority_interest.latest
    fcf = F.free_cash_flow.latest
    earnings = F.net_income_common_stockholders.latest
    total_yield = F.total_yield.latest
    
    columns = {
        'revenue_growth': revenue_growth,
        'earnings_growth': earnings_growth,
        'bps_growth': bps_growth,
        'dps_growth': dps_growth,
        'net_margin': net_margin,
        'roe': roe,
        'roic': roic,
        'cash': cash / BILLION,
        'op_income': op_income / BILLION,
        'debt_equity': debt_equity,
        'current_ratio': current_ratio,
        'current_assets': current_assets / BILLION,
        'total_debt': total_debt / BILLION,
        'total_liabilities': total_liabilities / BILLION,
        'earnings': earnings / BILLION,
        'fcf': fcf / BILLION,
        'total_yield': total_yield,
    }
    
    pipe = Pipeline(
        columns=columns,
        screen=mask
    )
    return pipe

today = datetime.now().date() - timedelta(days=1)
result = run_pipeline(make_pipeline(), today, today)

result
```
